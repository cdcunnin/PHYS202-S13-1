```python
import numpy.random
#define the road; length and number of cars
def Road(L,N):
    velocities = zeros(N,int)
    for CarMinusOne in range(0,N):
        velocities[CarMinusOne] = randint(0, L-N)
    cars = zeros(N,int)
    for i in range(N):
        cars[i] = i + 1
    road = zeros(L,int)
    car = 1
    position = zeros(N,int)
    while sum(road) != sum(cars):
        i = randint(0,L)
        if road[i] == 0:
            road[i] = car
            position[car-1] = i
            car +=1
    return position, cars, road, velocities

def GetGSubI(road,car,position):
    L = len(road)
    x = position[car-1] + 1
    while x < L and road[x] == 0:
        x += 1
    g = x - position[car-1]
    if x == L:
        x = 0
        while road[x] == 0:
            x += 1    
        g = x + L - position[car-1]
    return g

def STCA(road,cars,position,velocities,probability):
    N = len(cars)
    L = len(road)
    for CarMinusOne in range(0,N):
        g = GetGSubI(road,CarMinusOne + 1,position)
        if velocities[CarMinusOne] > g:
            velocities[CarMinusOne] = g
        elif velocities[CarMinusOne] < g:
            velocities[CarMinusOne] += 1
        if rand() <= probability and velocities[CarMinusOne] > 0:
            velocities[CarMinusOne] -= 1     
    for CarMinusOne in range(0,N):
        position[CarMinusOne] += velocities[CarMinusOne]
        if position[CarMinusOne] >= L:
            position[CarMinusOne] -= L
    road = zeros(L,int)
    for CarMinusOne in range(0,N):
        road[position[CarMinusOne]] = CarMinusOne + 1
    return road, cars, position, velocities        
            
def ASEP(road,cars,position,velocities):
    N = len(cars)
    L = len(road)
    Car = randint(0,N)
    for CarMinusOne in range(0,N):
        g = GetGSubI(road,CarMinusOne + 1,position)
        if velocities[CarMinusOne] > g:
            velocities[CarMinusOne] = g
        elif velocities[CarMinusOne] < g:
            velocities[CarMinusOne] += 1
    position[Car - 1] += velocities[Car -1]
    if position[Car -1] >= L:
        position[Car -1] -= L
    road = zeros(L,int)
    for CarMinusOne in range(0,N):
        road[position[CarMinusOne]] = CarMinusOne + 1
    return road, cars, position, velocities             
            
    
```

```python
def RunTimeSTCA(L,N,Runs):
    P,C,R,V = Road(L,N)
    Vavg = zeros(Runs, float)
    Positions = zeros((Runs,len(P)), int )
    Current = zeros(Runs, float)
    Pr = .3
    for i in range(Runs):
        R,C,P,V = STCA(R,C,P,V,Pr)
        Current[i] = float(sum(V)) / float(Runs)
        Vavg[i] = float(sum(V)) / float(len(V))
        Positions[i] = P
    return Current, Vavg, Positions

def RunTimeASEP(L,N,Runs):
    P,C,R,V = Road(L,N)
    Vavg = zeros(Runs, float)
    Positions = zeros((Runs,len(P)), int )
    Current = zeros(Runs, float)
    for i in range(Runs):
        R,C,P,V = ASEP(R,C,P,V)
        Current[i] = float(sum(V)) / float(Runs)
        Vavg[i] = float(sum(V)) / float(len(V))
        Positions[i] = P
    return Current, Vavg, Positions

        
```

```python
N = 10
L = 50
Runs = 150
CurrentS,VavgS,PositionS = RunTimeSTCA(L,N,Runs)
CurrentA,VavgA,PositionA = RunTimeASEP(L,N,Runs)
t = arange(0,150,1)
plot(t, CurrentS, label = "STCA current")
legend(loc='best')
xlabel("time(s)")
show()
plot(t, VavgS)
show()
plot(t, PositionS)
show()
plot(t, CurrentA)
show()
plot(t, VavgA)
show()
plot(t, PositionA)
show()
```
