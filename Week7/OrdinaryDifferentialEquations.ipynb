```python
g = 9.8
l = .5
th0 = 0
th01 = pi/4
th02 = pi/2
th03 = 3* pi /4
th04 = pi
t = arange(0.,10.+0.01,0.01)
theta_app = lambda t: th0*cos(sqrt(g/l)*t)
theta_app1 = lambda t: th01*cos(sqrt(g/l)*t)
theta_app2 = lambda t: th02*cos(sqrt(g/l)*t)
theta_app3 = lambda t: th03*cos(sqrt(g/l)*t)    
theta_app4 = lambda t: th04*cos(sqrt(g/l)*t)    
plot(t,theta_app(t),label='1')
plot(t,theta_app1(t),label='2')
plot(t,theta_app2(t),label='3')
plot(t,theta_app3(t),label='4')
plot(t,theta_app4(t),label='5')
legend(loc=3)
show()
    
```

When the initial angle is 0, the amplitude is at 0 and there is no motion when the pendulum remains at the starting point. When the starting angle is increased to $\pi/4$ and then to $3\pi/4$, the amplitude of the trajectory increases gradually. When the initial angle is $\pi$ the amplitude of the trajectory is at a maximum since the pendulum is as far away from the equilibrium point as possible.
