Using Matplotlib to make pretty graphs

```python
#simple line plot
t = arange(0.0,1.01,.01)
c = cos(4*pi*t)
s = sin(4*pi*t)
```

```python
print s
```

```python
print t
```

```python
plot(t,s)
```

```python
plot(t,s,color="blue",linewidth=2,label="$sin(4{\pi}t)$")
plot(t,c,color="red",linewidth=3,label="$cos(4{\pi}t)$")
xlabel("time (s)",fontsize=15)
ylabel("voltage (mV)",fontsize=15)
title("Super Sweet Graph",fontsize=20) 
legend() #to show label
grid(True)
savefig("simple_plot.pdf")
show()
```
