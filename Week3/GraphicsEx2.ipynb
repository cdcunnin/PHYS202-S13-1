```python
x = arange(-pi,pi,.01)
U = -cosh(x)**-2
i = 0 +1j
l =[0.0, 0.5, 1.0, 2.0]
z = ['r-','r--','r-.','r:']
g= 0
for k in l:
    b = ((i*k - tanh(x)) / (i*k +1)) * exp(i*k*x)
    plot (x,b*b.conjugate(), z[g])
    g += 1
plot(x,U) 
grid(True)
xlabel("x(rad)",fontsize=15)
```
