\begin{equation}

\end{equation}

\begin{equation}
\Omega_s = \omega \frac{sin\alpha}{sin\theta} = \frac{l}{\lambda_1} = \omega \frac {\sqrt{\lambda_3^2 + (\lambda_1^2 - \lambda_3^2) sin\alpha}}{\lambda_1}
\end{equation}

```python
x = arange(-pi,pi,.01)
U = -cosh(x)**-2
i = 0 +1j
l =[0.0, 0.5, 1.0, 2.0]
z = ['r-','r--','r-.','r:']
g= 0
plot(x,U, label ="$U(x)$")
for k in l:
    b = ((i*k - tanh(x)) / (i*k +1)) * exp(i*k*x)
    plot (x,b*b.conjugate(), z[g], label="$\psi^2,k%s$"%l)
    g += 1
    
 
grid(True)
xlabel("$x(rad)$")
ylabel("$U(x)$, $\psi$$^2(x)$")
legend(loc=3, prop={'size':7})
xticks(arange(-pi, 3*pi/2, pi/2),('$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$', '$+3\pi/2$'))
show()
```
