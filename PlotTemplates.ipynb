Line Plots
=


```python
y = np.arange(-5.0,5.0,.01) #set range of y values in array using arange
E = cosh(y) #set E to cosh values
p = sinh(y) #set p to sinh values
```

```python
#graphing this data
plot(y,E,color= "black", linestyle="-.", linewidth=3, label = "$E$ $=$ $cosh(y)$") #setting what the lines look like for E
plot(y,p,color= "blue", linestyle=":", linewidth=3, label = "$|$ $p$ $|$") #setting what the lines look like for p
title("Sample Plot Skeleton") #add in a title
xlabel('$X-axis$') #insert label for x-axis
ylabel('$Y-axis$') #insert label for y-axis
legend(loc=3) #location of legend, 3=lower left. can look up others or put location in ""
grid(True) #tuen on grid, can use 1 instead of True
ax = gca()
ax.spines['right'].set_color('none') #turn on lines (spines) for center and x and y axis 
ax.spines['left'].set_color('none') #turn on spines
ax.spines['top'].set_color('none') #turn on spines
ax.spines['bottom'].set_color('none') #turn on spines
axhline(y=0, color="black") #start x axis at zero instead of having zero at center ; when y=0
axvline(x=0,color="black") #turn on vertical axis
show()


```
