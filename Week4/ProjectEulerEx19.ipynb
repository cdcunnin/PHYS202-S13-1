```python
def days_in_month(month, year):
    if (month == 1) or (month == 3) or (month == 5) or (month == 7) or (month == 8) or (month == 10) or (month == 12):
        days = 31
        return days
    elif (month == 4) or (month == 6) or (month == 9) or (month == 11):
        days = 30
        return days
    elif (month == 2):
        if year % 4 == 0:
            days = 29
            return days
        else:
            days = 28
            return days
         
    else:
        print "Not Valid"
         
Week = {1:'Sunday', 2:'Monday', 3:'Tuesday', 4:'Wednesday', 5:'Thursday', 6:'Friday', 7:'Saturday'}
 
def sunday_count():
    count = 1 
    total = 0
    for year in range(1900, 2001):
        print "\nYear = ", year 
        for month in range(1, 13):
            D = days_in_month(month, year)
             
            for i in range(1, (D + 1)):
                count = count + 1       
                if (i == 1) and Week[count] == "Sunday":
                    print "%sst Week= %s " \
                          % (i, Week[count])
                    total = total + 1
                if count >= 7:
                    count = 0
    print "\nTotal Sundays = ", (total - 2)
sunday_count()
```
