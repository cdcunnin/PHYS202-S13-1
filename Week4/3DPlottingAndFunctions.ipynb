Electric Dipole Equation 
=

\begin{equation}
V(x,y) = \frac{kq}{\sqrt{x^2 + (y-d)^2}} - \frac{kq}{\sqrt{x^2 + (y+d)^2}}
\end{equation}

```python
#make x-y grid to cover the plane x and y b/n -2 and 2
dx = 0.06
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
print x.ndim
print len(x)

X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
# make function V(x,y)
k = 1; q = 1; d= 1
V = (k*q/(X**2 + (Y - d)**2)**(1/2.)) - (k*q/(X**2 + (Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d') #use these three lines for 3d graphs (from figure, get axis)
#help(Axes3D) #shows how stride and stuff change your graph
p = ax.plot_surface(X,Y,V, rstride=1, cstride=1, linewidth=0, cmap=cm.terrain) #hit tab after cm. to get list of colors
cb = fig.colorbar(p,shrink=.5)
show()
```

```python
#3D data represented as contours
levels = array([8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8]) #divide by two each time...
contour(X,Y,V, levels, linewidths=5,cmap=cm.gist_rainbow)
show()
```
